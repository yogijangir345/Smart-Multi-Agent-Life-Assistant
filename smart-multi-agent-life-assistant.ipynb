````python
class PlannerAgent:
    
    def create_plan(self, user_goal):
        
        plan = [
            f"Understand the goal: {user_goal}",
            "Break the task into smaller steps",
            "Send the task to the Tool Agent",
            "Store important information in Memory",
            "Ask Security Agent to validate the response",
            "Return the final answer to the user"
        ]
        
        return plan
````

````python
class PlannerAgent:

    def create_plan(self, user_goal):

        goal = user_goal.lower()

        if "trip" in goal or "travel" in goal:
            plan = [
                "Find destination details",
                "Create travel itinerary",
                "Estimate budget",
                "Recommend places to visit",
                "Generate final travel plan"
            ]

        elif "study" in goal:
            plan = [
                "Identify subjects",
                "Create study timetable",
                "Allocate revision time",
                "Generate study plan"
            ]

        else:
            plan = [
                "Understand the user's request",
                "Break the task into smaller steps",
                "Generate the final response"
            ]

        return plan
````

````python
planner = PlannerAgent()

goal = "Plan a 3-day trip to Delhi under ₹5000"

plan = planner.create_plan(goal)

print("User Goal:")
print(goal)

print("\nGenerated Plan:")

for i, step in enumerate(plan, start=1):
    print(f"{i}. {step}")
````

````python
class ToolAgent:

    def get_trip_details(self):

        trip_info = {
            "Destination": "Delhi",
            "Budget": "₹5000",
            "Places": [
                "India Gate",
                "Qutub Minar",
                "Red Fort",
                "Lotus Temple"
            ]
        }

        return trip_info
````

````python
tool = ToolAgent()

trip = tool.get_trip_details()

print("Trip Details")

for key, value in trip.items():
    print(f"{key}: {value}")
````

````python
planner = PlannerAgent()
tool = ToolAgent()

goal = "Plan a 3-day trip to Delhi under ₹5000"

plan = planner.create_plan(goal)
trip = tool.get_trip_details()

print("User Goal:")
print(goal)

print("\nPlanner Agent Output:")
for i, step in enumerate(plan, start=1):
    print(f"{i}. {step}")

print("\nTool Agent Output:")
for key, value in trip.items():
    print(f"{key}: {value}")
````

````python
class MemoryAgent:

    def __init__(self):
        self.history = []

    def save_goal(self, goal):
        self.history.append(goal)

    def show_history(self):
        return self.history
````

````python
memory = MemoryAgent()

memory.save_goal("Plan a 3-day trip to Delhi under ₹5000")
memory.save_goal("Create a 7-day study schedule")

print("Saved Goals:")

for i, goal in enumerate(memory.show_history(), start=1):
    print(f"{i}. {goal}")
````

````python
class SecurityAgent:

    def check_request(self, user_input):

        blocked_words = [
            "hack",
            "virus",
            "malware",
            "password"
        ]

        text = user_input.lower()

        for word in blocked_words:
            if word in text:
                return False

        return True
````

````python
security = SecurityAgent()

safe_request = "Plan a 3-day trip to Delhi"
unsafe_request = "How to hack a website"

print("Safe Request:")
print(security.check_request(safe_request))

print("\nUnsafe Request:")
print(security.check_request(unsafe_request))
````

````python
class OrchestratorAgent:

    def __init__(self, planner, tool, memory, security):
        self.planner = planner
        self.tool = tool
        self.memory = memory
        self.security = security

    def run(self, user_goal):

        if not self.security.check_request(user_goal):
            return "Request blocked by Security Agent."

        self.memory.save_goal(user_goal)

        plan = self.planner.create_plan(user_goal)

        trip = self.tool.get_trip_details()

        return {
            "goal": user_goal,
            "plan": plan,
            "trip": trip,
            "history": self.memory.show_history()
        }
    
````

````python
planner = PlannerAgent()
tool = ToolAgent()
memory = MemoryAgent()
security = SecurityAgent()

assistant = OrchestratorAgent(
    planner,
    tool,
    memory,
    security
)

result = assistant.run("Plan a 3-day trip to Delhi under ₹5000")

print("User Goal:")
print(result["goal"])

print("\nGenerated Plan:")
for i, step in enumerate(result["plan"], start=1):
    print(f"{i}. {step}")

print("\nTrip Information:")
for key, value in result["trip"].items():
    print(f"{key}: {value}")

print("\nMemory:")
for item in result["history"]:
    print("-", item)

````

````python
from google.adk.agents import LlmAgent, SequentialAgent
from google.adk.apps.app import App
from google.adk.runners import Runner
from google.adk.sessions import InMemorySessionService
from google.genai import types

print("Imports Successful ✅")
````

````python
security_agent = LlmAgent(
    name="security_agent",
    model="gemini-2.5-flash",
    instruction="Check if the user's request is safe."
)

planner_agent = LlmAgent(
    name="planner_agent",
    model="gemini-2.5-flash",
    instruction="Create a travel plan."
)

budget_agent = LlmAgent(
    name="budget_agent",
    model="gemini-2.5-flash",
    instruction="Estimate the travel budget."
)

travel_agent = LlmAgent(
    name="travel_agent",
    model="gemini-2.5-flash",
    instruction="Suggest places to visit."
)

print("LLM Agents Created ✅")
````

````python
life_assistant = SequentialAgent(
    name="life_assistant",
    sub_agents=[
        security_agent,
        planner_agent,
        budget_agent,
        travel_agent,
    ],
)

print("Sequential Agent Created ✅")
````

````python
app = App(
    name="life_assistant_app",
    root_agent=life_assistant
)

print("App Created ✅")
````

````python
session_service = InMemorySessionService()

runner = Runner(
    app=app,
    session_service=session_service
)

print("Runner Created ✅")
````

````python
session = await session_service.create_session(
    app_name=app.name,
    user_id="user_1"
)

print("Session Created ✅")
print("Session ID:", session.id)
````

````python
from google.genai import types

message = types.Content(
    role="user",
    parts=[
        types.Part(text="Plan a 3-day trip to Delhi under ₹5000")
    ]
)

for event in runner.run(
    user_id="user_1",
    session_id=session.id,
    new_message=message,
):
    print(event)
````

````python
def validate_request(user_input: str):
    """
    Basic security guardrail for user requests.
    """

    blocked_words = [
        "hack",
        "malware",
        "virus",
        "password",
        "credit card",
        "phishing",
        "exploit"
    ]

    text = user_input.lower()

    for word in blocked_words:
        if word in text:
            return False, f"❌ Blocked request: '{word}' detected."

    return True, "✅ Request is safe."
````

````python
print(validate_request("Plan a 3-day trip to Delhi"))
print(validate_request("How to hack a website"))
print(validate_request("Show my credit card password"))
````

````python
def secure_run(user_input):

    is_safe, message = validate_request(user_input)

    if not is_safe:
        print(message)
        return

    print("Security Check Passed ✅")
    print("Agent can now process the request.")
````

````python
secure_run("Plan a trip to Jaipur")
print("-"*40)
secure_run("How to hack Facebook")
````

````python
def run_secure_agent(user_input):

    # Security Check
    is_safe, message = validate_request(user_input)

    if not is_safe:
        print(message)
        return

    print("✅ Security Passed\n")

    message = types.Content(
        role="user",
        parts=[types.Part(text=user_input)]
    )

    for event in runner.run(
        user_id="user_1",
        session_id=session.id,
        new_message=message,
    ):
        print(event)
````

````python
run_secure_agent("Plan a 3-day trip to Delhi under ₹5000")
````

````python
run_secure_agent("How to hack a website")
````

````python
def evaluate_response(response_text):

    if len(response_text) < 50:
        return "⚠️ Response is too short."

    if "sorry" in response_text.lower():
        return "⚠️ Response may be incomplete."

    return "✅ Response Quality: Good"
````

````python
print(evaluate_response("Delhi is a great place to visit."))
print(
    evaluate_response(
        """Delhi offers a great mix of history, food, culture,
        and tourist attractions. You can comfortably explore
        the city in three days while staying within your budget."""
    )
)
````

````python
def secure_and_evaluate(user_input):

    # Step 1: Security Check
    is_safe, message = validate_request(user_input)

    if not is_safe:
        print(message)
        return

    print("✅ Security Passed\n")

    # Step 2: Run Agent
    message = types.Content(
        role="user",
        parts=[types.Part(text=user_input)]
    )

    final_response = ""

    for event in runner.run(
        user_id="user_1",
        session_id=session.id,
        new_message=message,
    ):

        print(event)

        if event.content:
            for part in event.content.parts:
                if hasattr(part, "text") and part.text:
                    final_response += part.text

    # Step 3: Evaluate
    print("\n")
    print(evaluate_response(final_response))
````

````python
secure_and_evaluate("Plan a 3-day trip to Delhi under ₹5000")
````

````python
secure_and_evaluate("How to hack a bank")
````

````python
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

logging.info("Production Logging Enabled ✅")
````

````python
import time

def monitor_execution(user_input):

    start_time = time.time()

    print("🚀 Starting Agent...\n")

    run_secure_agent(user_input)

    end_time = time.time()

    execution_time = end_time - start_time

    print("\n----------------------------")
    print(f"⏱ Execution Time: {execution_time:.2f} seconds")
````

````python
monitor_execution("Plan a 3-day trip to Delhi under ₹5000")
````

````python
def production_run(user_input):

    try:
        logging.info("New request received.")

        monitor_execution(user_input)

        logging.info("Request completed successfully.")

    except Exception as e:
        logging.error(f"Application Error: {e}")

        print("❌ Something went wrong.")
````
